```python
im1=Image.open("/Users/vijitchandna/Documents/ImageSet 2/blue.png")
im2=Image.open("/Users/vijitchandna/Documents/ImageSet 2/red.png")
print("image 1 ",im1.format, im1.size, im1.mode)

print("image 2",im2.format, im2.size, im2.mode)
```

```python
def merge(im1, im2):
    w = im1.size[0] + im2.size[0]
    h = max(im1.size[1], im2.size[1])
    newImage= Image.new("RGBA", (w, h))

    newImage.paste(im1)
    newImage.paste(im2, (im1.size[0], 0))

    return newImage


added=merge(im1,im2)
```
